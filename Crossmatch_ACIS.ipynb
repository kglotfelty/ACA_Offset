```python
import os
```

```python
mydat = open("match_no_tg.dat","r").readlines()
```

```python
mydat[1:3]
```

```python
from ciao_contrib.runtool import wcs_match

for line in mydat[1:-1]: # skip 1st and last lines

    master = line.split(' ')[0]
    master_id = master.split("_")[0][1:]

    xmatch = line.split('{')[1].split('}')[0].strip()
    for xx in xmatch.split(","):
        xx_id = xx.split("_")[0][1:]

        f = "Wavdetect/{a}/wavdetect/{a}_baseline.src"
        m = f.format(a=master_id)
        x = f.format(a=xx_id)

        if not os.path.exists(m) or not os.path.exists(x):
            #print("{} x {} -- skipped".format(master_id,xx_id))
            continue
            
        print("{} x {}".format(master_id,xx_id))

        os.makedirs("XMatch",exist_ok=True)
        outfile = "XMatch/{}_{}.xmatch".format(master_id,xx_id)

        wcs_match.punlearn()

        try:
            wcs_match(m,x,outfile,radius=2.0, clobber=True, method="trans")
        except:
            print("{} x {} -- failed".format(master_id,xx_id))
            
```
